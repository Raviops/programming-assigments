**7.** How many people need to be in a group so that is it more probable than not that at least two of them have birthdays on the same day of the year? Assume for simplicity that there are 365 days in a year and that all the 365 days are equally probable birthdays for a randomly chosen person.

_P(meeting someone with the same birthday) = 1/365 for one person, for two persons it will be 1/182.5 + 1/365. Twice the chance + the chance that these two persons have the same birthday. That means that for a number of people, x, the chance that within x there are two persons with the same birthday is 1/(365/x) + (x!/(x!-2)!)*(1/365). The outcome has to be 1.
